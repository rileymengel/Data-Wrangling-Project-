```python
import seaborn as sns

# Correlation matrix
correlation_matrix = clean_df[['CPI', 'rating', 'popularity_score']].corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap: CPI, Ratings, and Popularity')
plt.show()
```

```python
import seaborn as sns

# Filter for a specific genre (e.g., "Action")
action_movies = clean_df[clean_df['genres'].str.contains('Action', na=False)]

# Scatter plot with regression line
plt.figure(figsize=(8, 6))
sns.regplot(x='CPI', y='popularity_score', data=action_movies, scatter_kws={'alpha':0.6}, line_kws={"color": "red"})
plt.title('CPI vs Popularity for Action Movies')
plt.xlabel('CPI')
plt.ylabel('Popularity Score')
plt.show()
```
