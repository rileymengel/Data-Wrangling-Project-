```python
import pandas as pd
df=pd.read_csv('top_250_raw.csv')
```

```python
display(df.head())
```

```python
df[['rank', 'title_clean']] = df['title'].str.split('.', n=1, expand=True)

# Remove any leading/trailing spaces in the 'title_clean' column
df['title_clean'] = df['title_clean'].str.strip()

# Show the updated DataFrame
display(df.head())
```

```python
# Drop the original 'title' column
df = df.drop(columns=['title'])

# Rename 'title_clean' to 'title' for clarity
df = df.rename(columns={'title_clean': 'title'})

# Reorder the columns so 'rank' and 'title' come first
column_order = ['rank', 'title'] + [col for col in df.columns if col not in ['rank', 'title']]
df = df[column_order]

# Show the updated DataFrame
display(df.head())
```

```python
# Separate date in 2 columns "month" and "year"
df['month'] = df['release_date'].str.split(' ').str[0]  # Extract the month
df['year'] = df['release_date'].str.split(' ').str[-1]  # Extract the year

# Drop the original release_date column
df = df.drop(columns=['release_date'])

# Reorder the columns to include month and year at the end
column_order = [col for col in df.columns if col not in ['month', 'year']] + ['month', 'year']
df = df[column_order]

# Show the updated DataFrame
display(df.head())
```

```python
# Create new csv for cleaned movie data
df.to_csv('movies_cleaned.csv',index=False, encoding='utf-8')
```
